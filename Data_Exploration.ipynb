```python
#Imports
import pandas as pd
```

```python
# Load the dataset
file_path = 'ABT_GroupI.xlsx'
data = pd.read_excel(file_path)

# Display the first few rows of the dataset
data.head()
```

```python
# Calculate mean, median, and mode for numerical columns
mean_values = data.mean(numeric_only=True)
median_values = data.median(numeric_only=True)
mode_values = data.mode(numeric_only=True).iloc[0]

# Display the results
print("Mean:\n", mean_values)
print("\nMedian:\n", median_values)
print("\nMode:\n", mode_values)
```

```python
# Check for null values and count them for each column
null_counts = data.isnull().sum()

# Display the null counts
print(null_counts)
```

```python
# Find the 10 highest spending customers
top_spenders = data.sort_values(by='Amount_of_money', ascending=False).head(10)

# Display the result
print(top_spenders)
```

```python
average_age_top_spenders = top_spenders['Age'].mean()
print("Average Age of Top Spenders:", average_age_top_spenders)
```
